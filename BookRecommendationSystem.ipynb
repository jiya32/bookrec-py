Popularity Based Recommendation System

```python
#merging ratings and books
ratings_with_name = ratings.merge(books, on='ISBN')
```

```python
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'}, inplace=True)
num_rating_df
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_ratings'}, inplace=True)
avg_rating_df
```

```python
popularity_dataframe = num_rating_df.merge(avg_rating_df, on='Book-Title')
popularity_dataframe
```

```python
popularity_dataframe = popularity_dataframe[popularity_dataframe['num_ratings']>=250].sort_values('avg_ratings', ascending=False).head(50)
```

```python
popularity_dataframe = popularity_dataframe.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','num_ratings','avg_ratings']]
popularity_dataframe
```

Collaborative Filtering Based Recommender System

```python
X = ratings_with_name.groupby('User-ID').count()['Book-Rating']>200
active_users = X[X].index
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(active_users)]
```

```python
Y= filtered_rating.groupby('Book-Title').count()['Book-Rating'] >=10
fam_books = Y[Y].index
```

```python
final_rating= filtered_rating[filtered_rating['Book-Title'].isin(fam_books)]
```

```python
pt = final_rating.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
```

```python
pt.fillna(0, inplace=True)
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarityscore = cosine_similarity(pt)
similarityscore.shape
```

```python
def recommend(book_name):
  #index fetch
  index = np.where(pt.index==book_name)[0][0]
  similar_items = sorted(list(enumerate(similarityscore[index])), reverse=True, key=lambda x:x[1])[1:6]

  data=[]
  for i in similar_items:
    item=[]
    temp_df = books[books['Book-Title']==pt.index[i[0]]]
    temp_df.drop_duplicates('Book-Title')
    item.extend (temp_df.drop_duplicates('Book-Title')['Book-Title'].values)
    item.extend (temp_df.drop_duplicates('Book-Title')['Book-Author'].values)
    item.extend (temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values)

    data.append(item)

  return data
```
